```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
def read_data(filename):
    X = []
    Y = []
    with open(filename, 'r') as file:
        dataset = csv.reader(file)
        for row in dataset:
            X.append(row[1:])
            Y.append([row[0]])
        X = np.array(X).astype(float)
        Y = np.array(Y).astype(int)
    return (X, Y)
```

```python
(X, Y) = read_data('wine.data')
```

```python
C = np.cov(X.T)
(V, D) = np.linalg.eigh(C)
V = V[::-1]
W = V / np.sum(V)*100
D = D.T[::-1].T
print(W)
plt.plot(V,"bo")
plt.show()
```

```python
P = D[:, :2]
proj = X @ P

x = proj[:, 0]
y = proj[:, 1]

class1 = np.where(Y == 1)[0]
class2 = np.where(Y == 2)[0]
class3 = np.where(Y == 3)[0]

print('Red eigenvector: 1st principal component')
print('Green eigenvector: 2nd principal component')
plt.figure(figsize=(10, 10))
plt.quiver(np.mean(x), np.mean(y), P[:, 0], P[:, 1], color=['r', 'g'], scale=10)
plt.scatter(x[class1], y[class1], marker='1', label='Class 1')
plt.scatter(x[class2], y[class2], marker='2', label='Class 2')
plt.scatter(x[class3], y[class3], marker='3', label='Class 3')
plt.legend()
plt.grid()
plt.show()
```
